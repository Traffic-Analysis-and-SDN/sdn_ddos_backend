```python
import pandas as pd

# Define the path to your Parquet file
file_path = r'C:/Users/COOL/sdn_ddos_backend/dataset/DNS-testing.parquet'
```

```python
# Load the data from the Parquet file into a DataFrame
df = pd.read_parquet(file_path)
```

```python
# --- Perform Initial Analysis ---

# 1. Get a summary of the DataFrame, including data types and non-null values
print("--- Data Info ---")
df.info()
```

```python
df.describe()
```

```python
# 2. See all the feature (column) names available in the dataset
print("\n--- Column Names ---")
print(df.columns)
```

```python
# 3. Check the balance of BENIGN vs. DDOS traffic in the 'Label' column
print("\n--- Label Distribution ---")
print(df['Label'].value_counts())
```

```python
import numpy as np

# --- Step 3.1: Clean Column Names ---
# Good practice to remove any leading/trailing spaces from column names
df.columns = df.columns.str.strip()
```

```python
# --- Step 3.2: Select Features based on the Contract ---
# Create a list of the exact feature names from your finalized contract
# IMPORTANT: Don't forget to include the 'Label' column!
contract_features = [
    'Protocol',
    'Flow Duration',
    'Total Fwd Packets',
    'Total Backward Packets',
    'Fwd IAT Total',
    'Packet Length Mean',
    'FIN Flag Count',
    'SYN Flag Count',
    'Label' # The target variable for training
]

# Create a new DataFrame containing only the selected features
df_contract = df[contract_features].copy()
```

```python
# --- Step 3.3: Handle Missing or Infinite Data ---
# Replace any infinite values with NaN (Not a Number)
df_contract.replace([np.inf, -np.inf], np.nan, inplace=True)

# Drop any rows that still contain NaN values
df_contract.dropna(inplace=True)
```

```python
# --- Verification ---
# Display info on the new, clean DataFrame to confirm the changes
print("--- Cleaned DataFrame Info ---")
df_contract.info()

print(f"\nOriginal dataset shape: {df.shape}")
print(f"New dataset shape after cleaning and selection: {df_contract.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# --- Step 4.1: Separate Features (X) and Target (y) ---
# X contains all the feature columns from your contract
X = df_contract.drop('Label', axis=1)

# y contains only the 'Label' column
y = df_contract['Label']
```

```python
# --- Step 4.2: Encode the Labels ---
# Machine learning models require numerical inputs, so we convert the
# text labels ('Benign', 'DrDoS_DNS') into numbers (e.g., 0 and 1).
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# --- Step 4.3: Split Data into Training and Testing Sets ---
# This splits the data into a 70% training set and a 30% testing set.
# The 'random_state=42' ensures that the split is the same every time you run the code.
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.3, random_state=42)

```

```python
# --- Verification ---
print("Data has been split successfully.")
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# 1. Create an instance of the Random Forest model
# n_estimators=100 means the model will use 100 decision trees.
# random_state=42 ensures you get the same results every time you run it.
model = RandomForestClassifier(n_estimators=100, random_state=42)

# 2. Train the model using your training data
model.fit(X_train, y_train)

print("Model training is complete!")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Use the trained model to make predictions on the test data
predictions = model.predict(X_test)
```

```python
# 2. Calculate and print the performance metrics
print("--- Model Performance Metrics ---")

# Accuracy: How often the model is correct overall.
print(f"Accuracy: {accuracy_score(y_test, predictions):.4f}")

# Precision: Of all the flows the model labeled as attacks, how many were actually attacks.
print(f"Precision: {precision_score(y_test, predictions):.4f}")

# Recall: Of all the real attacks that occurred, how many did the model find.
print(f"Recall: {recall_score(y_test, predictions):.4f}")

# F1-Score: A weighted average of Precision and Recall.
print(f"F1-Score: {f1_score(y_test, predictions):.4f}")
```

```python
# 3. Generate and display the confusion matrix
print("\n--- Confusion Matrix ---")
cm = confusion_matrix(y_test, predictions)

# Use a heatmap for better visualization
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'DDoS'], yticklabels=['Benign', 'DDoS'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
import joblib
import os

# --- Final Step: Save the Model in a 'models' Folder ---

# Define the folder and the full file path
model_folder = r'C:/Users/COOL/sdn_ddos_backend/models'
file_path = os.path.join(model_folder, 'ddos_model.joblib')

# Create the 'models' directory if it doesn't exist
os.makedirs(model_folder, exist_ok=True)

# Save the trained model to the specified path
joblib.dump(model, file_path)

print(f"Model successfully saved to: {file_path}")
```
